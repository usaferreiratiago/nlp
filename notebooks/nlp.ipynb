```python
# Extract only numeric information from the provided CNPJ text
import re
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import ParagraphStyle, getSampleStyleSheet
from reportlab.lib import colors
from reportlab.lib.units import inch
from reportlab.lib.pagesizes import A4

# Original extracted text (from your PDF)
text = """
17.437.384/0001-07
19/01/2013
62.09-1-00
95.11-8-00
213-5
25
35.182-143
(31) 3847-1020
14/06/2021
24/12/2025 06:15:24
"""

# Regex to capture numeric patterns
patterns = [
    r'\d{2}\.\d{3}\.\d{3}/\d{4}-\d{2}',  # CNPJ
    r'\d{2}/\d{2}/\d{4}',                # Dates
    r'\d{2}\.\d{2}-\d-\d{2}',            # CNAE
    r'\d{3}-\d',                         # Natureza jurídica
    r'\d{5}',                            # Número simples
    r'\d{2}\.\d{3}-\d{3}',                # CEP
    r'\(\d{2}\)\s?\d{4,5}-\d{4}',        # Telefone
]

numbers_found = []

for pattern in patterns:
    matches = re.findall(pattern, text)
    numbers_found.extend(matches)

# Remove duplicates while preserving order
seen = set()
numbers_clean = []
for num in numbers_found:
    if num not in seen:
        numbers_clean.append(num)
        seen.add(num)

# Create new PDF with only numeric data
file_path = '/home/vboxuser/Downloads/CNPJ_numeros_extraidos.pdf'
doc = SimpleDocTemplate(file_path, pagesize=A4)
elements = []

styles = getSampleStyleSheet()
style = styles["Normal"]

elements.append(Paragraph("<b>Informações Numéricas Extraídas</b>", styles["Heading1"]))
elements.append(Spacer(1, 0.3 * inch))

for num in numbers_clean:
    elements.append(Paragraph(num, style))
    elements.append(Spacer(1, 0.2 * inch))

doc.build(elements)

file_path
```

```python
# Extract only numeric information from the provided CNPJ text
import re
import csv

# Original extracted text (from your PDF)
text = """
17.437.384/0001-07
19/01/2013
62.09-1-00
95.11-8-00
213-5
25
35.182-143
(31) 3847-1020
14/06/2021
24/12/2025 06:15:24
"""

# Regex patterns with labels
patterns = {
    "CNPJ": r'\d{2}\.\d{3}\.\d{3}/\d{4}-\d{2}',
    "DATE": r'\d{2}/\d{2}/\d{4}',
    "CNAE": r'\d{2}\.\d{2}-\d-\d{2}',
    "NATUREZA_JURIDICA": r'\d{3}-\d',
    "CEP": r'\d{2}\.\d{3}-\d{3}',
    "PHONE": r'\(\d{2}\)\s?\d{4,5}-\d{4}',
    "SIMPLE_NUMBER": r'\b\d{1,5}\b'
}

results = []

for label, pattern in patterns.items():
    matches = re.findall(pattern, text)
    for match in matches:
        results.append((label, match))

# Remove duplicates while preserving order
seen = set()
clean_results = []
for item in results:
    if item not in seen:
        clean_results.append(item)
        seen.add(item)

# Define CSV file path
file_path = "/home/vboxuser/Downloads/CNPJ_numeros_extraidos.csv"

# Write CSV
with open(file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["TYPE", "VALUE"])  # Header
    writer.writerows(clean_results)

print("CSV criado em:", file_path)
```
