```python
from pathlib import Path
from pypdf import PdfReader
from PyPDF2 import PdfReader

# Caminho da pasta Downloads
downloads_path = Path("/home/vboxuser/Downloads")

# Percorrer todos os PDFs da pasta
for pdf_file in downloads_path.glob("*.pdf"):
    print(f"\nProcessando: {pdf_file.name}")
    
    try:
        reader = PdfReader(pdf_file)
        text = ""
        
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
        
        print(text[:500])  # Mostra apenas os primeiros 500 caracteres
        
    except Exception as e:
        print(f"Erro ao processar {pdf_file.name}: {e}")
```

```python
import re
from pathlib import Path
from pypdf import PdfReader
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.units import inch
from reportlab.lib.pagesizes import A4

# Caminho da pasta Downloads
downloads_path = Path("/home/vboxuser/Downloads")

all_text = ""

# 1️⃣ Ler todos os PDFs da pasta
for pdf_file in downloads_path.glob("*.pdf"):
    print(f"Processando: {pdf_file.name}")
    
    try:
        reader = PdfReader(pdf_file)
        
        for page in reader.pages:
            all_text += page.extract_text() or ""
            
    except Exception as e:
        print(f"Erro ao processar {pdf_file.name}: {e}")

# 2️⃣ Capturar TODOS os números encontrados
# (qualquer sequência numérica com possíveis pontuações)
pattern = r'[\d\./\-\(\)\:]+'

numbers_found = re.findall(pattern, all_text)

# 3️⃣ Remover duplicados preservando ordem
seen = set()
numbers_clean = []

for num in numbers_found:
    if num not in seen and any(char.isdigit() for char in num):
        numbers_clean.append(num)
        seen.add(num)

# 4️⃣ Criar novo PDF apenas com números
file_path = '/home/vboxuser/Downloads/Numeros_extraidos_de_todos_PDFs.pdf'

doc = SimpleDocTemplate(file_path, pagesize=A4)
elements = []

styles = getSampleStyleSheet()

elements.append(Paragraph("<b>Números Extraídos dos PDFs</b>", styles["Heading1"]))
elements.append(Spacer(1, 0.3 * inch))

for num in numbers_clean:
    elements.append(Paragraph(num, styles["Normal"]))
    elements.append(Spacer(1, 0.2 * inch))

doc.build(elements)

print("PDF criado em:", file_path)
```

```python
# Extract only numeric information from the provided CNPJ text
import re
import csv

# Original extracted text (from your PDF)
text = """
17.437.384/0001-07
19/01/2013
62.09-1-00
95.11-8-00
213-5
25
35.182-143
(31) 3847-1020
14/06/2021
24/12/2025 06:15:24
"""

# Regex patterns with labels
patterns = {
    "CNPJ": r'\d{2}\.\d{3}\.\d{3}/\d{4}-\d{2}',
    "DATE": r'\d{2}/\d{2}/\d{4}',
    "CNAE": r'\d{2}\.\d{2}-\d-\d{2}',
    "NATUREZA_JURIDICA": r'\d{3}-\d',
    "CEP": r'\d{2}\.\d{3}-\d{3}',
    "PHONE": r'\(\d{2}\)\s?\d{4,5}-\d{4}',
    "SIMPLE_NUMBER": r'\b\d{1,5}\b'
}

results = []

for label, pattern in patterns.items():
    matches = re.findall(pattern, text)
    for match in matches:
        results.append((label, match))

# Remove duplicates while preserving order
seen = set()
clean_results = []
for item in results:
    if item not in seen:
        clean_results.append(item)
        seen.add(item)

# Define CSV file path
file_path = "/home/vboxuser/Downloads/CNPJ_numeros_extraidos.csv"

# Write CSV
with open(file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["TYPE", "VALUE"])  # Header
    writer.writerows(clean_results)

print("CSV criado em:", file_path)
```
