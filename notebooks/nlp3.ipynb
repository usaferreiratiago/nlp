```python
from PyPDF2 import PdfReader

with open('/home/vboxuser/Downloads/CVs/Data Engineer-Tiago H R Ferreira_eng.pdf', 'rb') as f:
    pdf_reader = PdfReader(f)
    
    print("Número de páginas:", len(pdf_reader.pages))
    
    texto_completo = ""
    for pagina in pdf_reader.pages:
        texto_completo += pagina.extract_text() + "\n"  # Adiciona uma quebra de linha entre páginas

print(texto_completo)
```

```python
from pdf2image import convert_from_path
import pytesseract

# Caminho do PDF
caminho_pdf = '/home/vboxuser/Downloads/CVs/Data Engineer-Tiago H R Ferreira_eng.pdf'

# Converte todas as páginas do PDF em imagens
pages = convert_from_path(caminho_pdf)

# Extrai texto de cada página usando OCR
all_text = ""
for i, page in enumerate(pages):
    text = pytesseract.image_to_string(page, lang='eng')  # 'eng' para inglês
    all_text += text + "\n"
    print(f"--- Texto da página {i+1} ---")
    print(text)

# Salvar todo o texto extraído em arquivo
with open('extracted_text.txt', 'w', encoding='utf-8') as f:
    f.write(all_text)
```

```python
import pdfplumber

caminho_pdf = '/home/vboxuser/Downloads/CVs/Data Engineer-Tiago H R Ferreira_eng.pdf'

with pdfplumber.open(caminho_pdf) as pdf:
    print("Número de páginas:", len(pdf.pages))
    
    texto_completo = ""
    for i, pagina in enumerate(pdf.pages):
        # Extrai o texto preservando o layout original (usa espaços para alinhamento)
        texto_pagina = pagina.extract_text(layout=True)
        if texto_pagina:
            # Adiciona um separador visível entre páginas (opcional)
            if i > 0:
                texto_completo += "\n" + "="*50 + "\n"
            texto_completo += texto_pagina

# Exibe o resultado
print(texto_completo)

# Opcional: salvar em um arquivo .txt para verificação
with open("cv_extraido.txt", "w", encoding="utf-8") as f:
    f.write(texto_completo)
```
