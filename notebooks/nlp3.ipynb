```python
from PyPDF2 import PdfReader

with open('/home/vboxuser/Downloads/CVs/Data_Lead_Tiago_H_R_Ferreira_eng.pdf', 'rb') as f:
    pdf_reader = PdfReader(f)
    
    # Mostra quantas páginas existem
    print("Número de páginas:", len(pdf_reader.pages))
    
    # Pega a primeira página
    page_two_text = pdf_reader.pages[0].extract_text()

print(page_two_text)
```

```python
from pdf2image import convert_from_path
import pytesseract

# Caminho do PDF
caminho_pdf = '/home/vboxuser/Downloads/CVs/Data_Lead_Tiago_H_R_Ferreira_eng.pdf'

# Converte todas as páginas do PDF em imagens
pages = convert_from_path(caminho_pdf)

# Extrai texto de cada página usando OCR
all_text = ""
for i, page in enumerate(pages):
    text = pytesseract.image_to_string(page, lang='eng')  # 'eng' para inglês
    all_text += text + "\n"
    print(f"--- Texto da página {i+1} ---")
    print(text)

# Salvar todo o texto extraído em arquivo
with open('extracted_text.txt', 'w', encoding='utf-8') as f:
    f.write(all_text)
```

```python
with open('contacts.txt','a+') as f:
  f.write(page_two_text)
  f.seek(0)
  print(f.read())
```

```python
import re

# Supondo que page_two_text contenha todo o texto da página
# Extrai todas as sequências de números
numeros_lista = re.findall(r'\d+', page_two_text)

# Salva cada sequência em uma linha no arquivo
with open('contacts.txt','a+') as f:
    for numero in numeros_lista:
        f.write(numero + '\n')  # cada número em uma linha
    f.seek(0)
    print(f.read())
```
